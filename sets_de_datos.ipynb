Elimina las filas con valores faltantes del dataset y guarda el resultado en una nueva variable

```python
datos_na = datos.dropna()
```

Muestra un resumen del nuevo dataset sin valores faltantes incluyendo columnas tipos de datos y cantidad de registros

```python
datos_na.info()
```

Accede a la columna ocean_proximity del dataset para visualizar sus valores categóricos


```python
#convertir la caracteristica categorica a numerica
#Proximidad al oceano
datos_na['ocean_proximity']
```

Cuenta la frecuencia de cada categoría en la columna ocean_proximity para ver cuántos valores hay de cada tipo

```python
datos_na['ocean_proximity'].value_counts()
#1,2,3,4,5
```

Convierte la columna categórica ocean_proximity en variables numéricas binarias usando one hot encoding

```python
#Dumies / One-Hot Encoding
# NEAR BAY  INLAND  NEAR OCEAN
#   1          0         0
#   0          1         0

dummies = pd.get_dummies(datos_na["ocean_proximity"], dtype=int)
```

Agrega al dataset las nuevas columnas generadas por el one hot encoding

```python
datos_na = datos_na.join(dummies)
```

Muestra las primeras filas del dataset actualizado con las nuevas columnas añadidas

```python
datos_na.head()
```

Elimina la columna ocean_proximity del dataset para dejar solo variables numéricas


```python
datos_na = datos_na.drop(["ocean_proximity"], axis=1)
```
