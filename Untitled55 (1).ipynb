**Aqui se importan las librerias y el dataset a utilizar**

```python
import pandas as pd
datos = pd.read_csv("housing.csv")
```

**Aqui se muestran los datos del dataset**

```python
datos.head()
```

**Aqui se muestra graficamente los datos del dataset**

```python
import seaborn as sb
sb.scatterplot(x="latitude", y="longitude", data = datos, hue="median_income",
               size=datos["median_income"])
```

**Aqui se muestran solo datos especificos del dataset**

```python
datos.loc[:, ["latitude", "longitude", "median_income"]]
```

**Aque se empieza a hace el modelo predictivo, se empieza a entrenar**

```python
from sklearn.cluster import KMeans

X = datos.loc[:, ["latitude", "longitude", "median_income"]]
modelo = KMeans(n_clusters=6)
predicciones = modelo.fit_predict(X)
```

```python
predicciones.shape
```

```python
X["segmento_economico"] = predicciones
```

```python
X
```

```python
X["segmento_economico"].value_counts()
```

**Aqui se muestra en un mapa por secciones el segmento economico**

```python
import seaborn as sb
sb.scatterplot(x="latitude", y="longitude", data = X, hue="segmento_economico",
               palette="bright")
```

**Aqui se muestra en una grafica de barras los segmentos economicos dejandote ver el conteco de dichos segmentos**

```python
sb.countplot(x="segmento_economico", data=X)
```

**Aqui se muestra el ingreso de cada segmento economico, viendo que 0 es el mayor**

```python
X.groupby(["segmento_economico"])["median_income"].mean()
```
