***LA ELABORACION DEL DICCIONARIO***
<br>
*EN ESTA CELDA SE ASIGNA LA TRADUCCION A LA PALABRA A TRADUCIR*

```python
dic = {
    "hola": "hello",
    "mundo": "world",
    "perro": "dog",
    "gato": "cat",
    "casa": "house",
    "comida": "food",
    "agua": "water",
    "escuela": "school",
    "amigo": "friend",
    "trabajo": "work"
}
```

**PROCESO DE TRADUCCION**
<BR>
*AQUI SE HACE EL PROCESO DE TRADUCCION*

```python
def translate(word):
  return dic.get(word, "no encontrada")
```

**IMPLEMENTACION**
<BR>
*AQUI SE PONE A PRUEBA EL MODELO DE TRADUCCION*

```python
word = "trabajo"
print(translate(word))
```
